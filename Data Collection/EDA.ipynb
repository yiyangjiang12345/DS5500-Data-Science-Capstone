## Analyze the review texts length for insights into user sentiment:

```python
# Review length
df['review_length'] = df['review'].apply(lambda x: len(str(x)))

# Distribution of review length
sns.histplot(data=df, x='review_length', bins=50)
plt.title('Review Length Distribution')
plt.show()
```

This suggests that most users tend to write brief reviews on Steam. There are very few long reviews, indicating that extended commentary is much less common among users.

## Investigate how user engagement relates to review numbers

```python
# Number of games owned vs. number of reviews written
sns.scatterplot(data=df, x='author_num_games_owned', y='author_num_reviews', hue='voted_up', alpha=0.5)
plt.title('Games Owned vs. Reviews Written')
plt.show()
```

The scatter plot indicates that most users own a relatively small number of games and write few reviews, as shown by the dense cluster of points near the origin. There is no strong visual correlation between the number of games owned and the number of reviews written, suggesting that owning more games does not necessarily lead to writing more reviews. The color differentiation for 'voted_up' shows that both positive and negative reviews are spread across users with varying numbers of games owned, without a distinct pattern favoring either sentiment.

## Language Distribution

```python
# Language distribution of reviews
df['language'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Review Languages')
plt.ylabel('Number of Reviews')
plt.show()
```

The bar chart indicates that English is the predominant language for reviews on the platform, followed by Chinese, with both languages significantly outpacing others. The distribution shows a steep drop-off in review counts after the first two languages, suggesting a more diverse but smaller user base for the remaining languages.

```python
# See how sentiment (positive/negative reviews) is distributed across different languages, 
# which might reveal language-specific trends
pd.crosstab(df['language'], df['voted_up']).plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Sentiment Distribution by Language')
plt.xlabel('Language')
plt.ylabel('Number of Reviews')
plt.show()
```

The bar chart suggests that, across various languages, the majority of reviews are positive (voted up), with English and Chinese having the highest number of positive reviews. Notably, the proportion of positive reviews in Chinese is overwhelmingly large compared to other languages, which may indicate cultural differences in review behavior or a higher level of satisfaction among Chinese-speaking users.

## Correlation Analysis

```python
# Focus on features that could influence sentiment
features = ['author_num_games_owned', 'author_playtime_forever', 'votes_up', 'votes_funny', 'weighted_vote_score', 'comment_count']
corr = df[features].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

The heatmap of the correlation matrix indicates that there are generally low to moderate correlations between the features presented. The highest positive correlation is observed between votes_up and comment_count, suggesting that reviews which receive more helpful votes also tend to have more comments. Additionally, most features do not show strong correlations with each other, indicating that no single feature strongly predicts another within this dataset.

```python
sns.countplot(x='written_during_early_access', hue='voted_up', data=df)
plt.title('Sentiment for Early Access Reviews')
plt.xlabel('Written During Early Access')
plt.ylabel('Count')
plt.legend(title='Voted Up', loc='upper right')
plt.show()
```

## Game Genre

```python
import pandas as pd
import ast

# Load the data
df = pd.read_csv('genres_by_game.csv')

# Convert the string representation of a list into an actual list
df['Top 3 Genres'] = df['Top 3 Genres'].apply(ast.literal_eval)

# Explode the 'Top 3 Genres' column to have one genre per row
df_exploded = df.explode('Top 3 Genres')
```

```python
df.head()
```

```python
# Determine the frequency of each genre across all games
import matplotlib.pyplot as plt

genre_counts = df_exploded['Top 3 Genres'].value_counts()
genre_counts.plot(kind='bar', figsize=(14, 7))
plt.title('Frequency of Genres')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Identify which genres commonly appear together
from itertools import combinations
from collections import Counter

# Find all combinations of genres for each game and count them
genre_combinations = Counter()

df['Top 3 Genres'].apply(lambda genres: genre_combinations.update(combinations(genres, 2)))

# Display the most common combinations
for combo, count in genre_combinations.most_common(10):
    print(f"{combo}: {count}")
```

The combination of 'Adventure' and 'Indie' is the most frequent, occurring 1300 times, suggesting that indie developers are particularly inclined to create adventure games, or that this combination is popular among Steam users.

'Action' paired with both 'Adventure' and 'Indie' are also highly common, reflecting a strong market or player interest in action-oriented gameplay within these genres.

There is notable interest in games that combine 'Simulation' with 'Indie' and 'Strategy', as well as those that merge RPG elements with 'Adventure', 'Action', or 'Indie' genres. This could indicate a player preference for games that offer rich storytelling or strategic depth, often found in RPGs and strategy games.

The prevalence of 'Indie' in many of the top pairs suggests that independent game development is a significant force on the platform, often intersecting with various genres to cater to diverse gaming preferences.

```python
# Add a column to indicate if a game is Early Access
df['Is Early Access'] = df['Top 3 Genres'].apply(lambda genres: 'Early Access' in genres)

early_access_counts = df['Is Early Access'].value_counts()
early_access_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Early Access Games')
plt.ylabel('')  # Hide the y-label
plt.show()
```
